### Introduction 

The data we'll use for the Fall 2024 STA130 course project is based on the [Canadian Social Connection Survey](https://casch.org/cscs) (CSCS) run by [genwell](https://genwell.ca/) and the [Canadian Alliance for Social Connection and Health](https://casch.org/) (CASCH). This data is used to produce [Rapid Evidence Briefs](https://casch.org/publications) which scientifically analyze the importance of social connection and community engagement -- relationships with family, friends, coworkers, neighbours, and strangers -- for personal health and well-being. The purpose of our course project is to help raise awareness about and interest in this often under appreciated topic. To do so STA130 students will do "data blitz" to discover interesting information and identify key statistics hidden inside the CSCS data. The findings each project team produces will be shared with other UofT courses (in Digital Humanities, Writing, Sociology, etc.) and other genwell and CASCH associated teams specializing in communications and media relations, and they will transform the most interesting findings of our teams into content for social media campaigns. If our findings are important enough, they may even form the basis of future Rapid Evidence Briefs...

> In conjunction with this project, we are taking part in a research project "investigating the impact of community engaged learning on student happiness, satisfaction, and well-being" (alongside multiple courses in Pharmacology and Toxicology, Human Biology, and Psychology). We are curious to know if this project gives you a sense of engagement and contribution to our broader Canadian community, and how this potential possibility to express your altruistic tendencies turn might influence your own experience in the course and in your own personal life. So, we're exploring data investigating "the importance of social connection and community engagement" while at the same time exploring the relevance and impact of this course project activity to our own sense of "connection and community engagement" and it's potential benefits to "personal health and well-being". 

To use this data ethically, professionally, and appropriately we must take a quick look and abide by the [data use agreement](https://static1.squarespace.com/static/60283c2e174c122f8ebe0f39/t/6239c284d610f76fed5a2e69/1647952517436/Data+Use+Agreement+for+the+Canadian+Social+Connection+Survey.pdf). You should also take a quick look at the [list of available variables](https://drive.google.com/file/d/1ISVymGn-WR1lcRs4psIym2N3or5onNBi/view) and then check out the [data](https://drive.google.com/file/d/1mbUQlMTrNYA7Ly5eImVRBn16Ehy9Lggo/view) itself (available at the bottom of the [CSCS](https://casch.org/cscs) webpage).

This project was originated and sourced [by Christine Ovcaric](https://www.linkedin.com/feed/update/urn:li:activity:7232457815112450048/) of UofT's [Experiential Learning Sandbox](https://www.artsci.utoronto.ca/faculty-staff/experiential-learning/curriculum-integrated/sandbox) program, who connected Prof. Schwartz [Dr. Kiffer Card](https://casch.org/kiffer-card)

- Etc., you get the point...

Dr. Card's objectives for this project are as introduced above, but this also offers STA130 students the opportunity to practice applying the skills and methods they're learning in STA130 on some real-world data that has the potential to actually matter. Summary statistics, visualizations, group comparisons, and associations and classification are all likely to be useful in the searching through the CSCS survey dataset to see what it tells us about social connection and well-being. As also described above, this project additionally gives STA130 students an opportunity to "give back" by contributing to a meaningful effort to understand key factors that contributed to improved personal health. STA130 students are likely to be able to intuitively understand and explore many of the variables as they work to identify key findings in the CSCS survey datasets and provide the initial analyses that could be the seeds of social media campaigns presenting findings that genwell wants to share with the public.

The evaluation of your own performance on the course project should be self-reflective and address what you did well and do not do well, and what went well and did not go well regarding teamwork within the group. 

On the day of the poster fair, you will be assigned a one-hour block of time to be at your poster with your group, and another one-hour block of time for you to circulate to examine and discuss the work of other project groups. In this time, you must visit at least two other groups to create your [evaluations](link).
