```python
import plotly.graph_objects as go
import pandas as pd

# Load the penguins dataset
penguins_url = "https://raw.githubusercontent.com/mwaskom/seaborn-data/master/penguins.csv"
penguins = pd.read_csv(penguins_url)

# Filter out rows with missing flipper_length_mm
penguins = penguins.dropna(subset=["flipper_length_mm"])

# Create a histogram for each species and add mean, median, range, IQR, and 2SD markers
species_list = penguins['species'].unique()
fig = go.Figure()

for species in species_list:
    species_data = penguins[penguins['species'] == species]
    flipper_length = species_data['flipper_length_mm']
    
    # Add histogram
    fig.add_trace(go.Histogram(x=flipper_length, name=species, opacity=0.75))

    # Compute statistics
    mean = flipper_length.mean()
    median = flipper_length.median()
    min_value = flipper_length.min()
    max_value = flipper_length.max()
    q1 = flipper_length.quantile(0.25)
    q3 = flipper_length.quantile(0.75)
    iqr = q3 - q1
    std_dev = flipper_length.std()

    # Add mean and median lines
    fig.add_vline(x=mean, line_width=2, line_dash="dash", line_color="blue", annotation_text=f"Mean {species}")
    fig.add_vline(x=median, line_width=2, line_dash="dash", line_color="green", annotation_text=f"Median {species}")

    # Add range (min to max)
    fig.add_vrect(x0=min_value, x1=max_value, line_width=0, fillcolor="red", opacity=0.1, annotation_text=f"Range {species}")

    # Add IQR (q1 to q3)
    fig.add_vrect(x0=q1, x1=q3, line_width=0, fillcolor="yellow", opacity=0.2, annotation_text=f"IQR {species}")

    # Add 2 standard deviations range (mean ± 2*std_dev)
    fig.add_vrect(x0=mean - 2 * std_dev, x1=mean + 2 * std_dev, line_width=0, fillcolor="purple", opacity=0.1, annotation_text=f"2SD {species}")

# Update layout
fig.update_layout(barmode='overlay', title="Histogram of Flipper Length (mm) by Species with Statistical Markers",
                  xaxis_title="Flipper Length (mm)", yaxis_title="Count", template="plotly_white")

# Show plot
fig.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Create a 1x3 subplot layout for the KDE plots
species_list = penguins['species'].unique()
fig, axes = plt.subplots(1, 3, figsize=(18, 6))

# Plot KDE for each species with statistical markers
for i, species in enumerate(species_list):
    species_data = penguins[penguins['species'] == species]
    flipper_length = species_data['flipper_length_mm']
    
    # KDE plot
    sns.kdeplot(flipper_length, ax=axes[i], label=f'{species} KDE')
    
    # Compute statistics
    mean = flipper_length.mean()
    median = flipper_length.median()
    min_value = flipper_length.min()
    max_value = flipper_length.max()
    q1 = flipper_length.quantile(0.25)
    q3 = flipper_length.quantile(0.75)
    iqr = q3 - q1
    std_dev = flipper_length.std()
    
    # Add mean and median lines
    axes[i].axvline(mean, color='blue', linestyle='--', label='Mean')
    axes[i].axvline(median, color='green', linestyle='--', label='Median')
    
    # Add range (min to max)
    axes[i].axvspan(min_value, max_value, color='red', alpha=0.1, label='Range')
    
    # Add IQR (q1 to q3)
    axes[i].axvspan(q1, q3, color='yellow', alpha=0.2, label='IQR')
    
    # Add 2 standard deviations range (mean ± 2*std_dev)
    axes[i].axvspan(mean - 2 * std_dev, mean + 2 * std_dev, color='purple', alpha=0.1, label='±2 Std Dev')
    
    # Set plot title and labels
    axes[i].set_title(f'KDE Plot for {species}')
    axes[i].set_xlabel('Flipper Length (mm)')
    axes[i].set_ylabel('Density')
    axes[i].legend()

# Adjust layout and display the plots
plt.tight_layout()
plt.show()
```

```python
Box Plots:
Description: Box plots (also called box-and-whisker plots) summarize data distributions by showing the median, quartiles (25th and 75th percentiles), and possible outliers. The "box" represents the interquartile range (IQR), which contains the middle 50% of the data. "Whiskers" extend to the smallest and largest data points within a certain range, usually 1.5 times the IQR.
    Histograms:
Description: Histograms are graphical representations of data distributions where the data is divided into bins, and the height of each bar represents the frequency of data points in that bin. They show the shape of the data distribution more explicitly than a box plot.
    Box Plots:
Pros:
Very compact and clean summary of the data.
Easily show quartiles, median, spread, and potential outliers.
Ideal for comparing multiple distributions side by side.
Cons:
Don't show the actual shape of the distribution (e.g., modality or skewness).
Outliers are just shown as points, without much context about their relative position.
Histograms:
Pros:
Easy to understand and visualize the shape of the data distribution.
Good for showing frequencies and how data is distributed across intervals.
Highlights skewness and multimodality well.
Cons:
The choice of bin width can significantly alter the visual result (too many bins may make the data look noisy, too few can oversimplify the distribution).
Not ideal for small datasets, where the histogram can look blocky or sparse.
Histograms would be my preference because they offer an intuitive balance between showing the shape of the data and providing a simple count of frequencies. For example, they visually highlight skewness, bimodality, and outliers much more explicitly than box plots. Additionally, they are easier to interpret than KDEs, especially for non-experts. Even though histograms depend on bin width, it’s fairly easy to adjust bins to balance detail and readability.
```

```python
explore the general relationship between the mean and median and "right" and "left" skewness (and why this is); what the following code does and how it works; and then explain (in your own words) the relationship between the mean and median and "right" and "left" skewness and what causes this, using and extending the code to demonstrate your explanation through a sequence of notebook cells.
ChatGPT ：
ChatGPT
To explore the relationship between the mean, median, and skewness (specifically right and left skewness), let’s break it down and review what skewness means, how it affects the mean and median, and then explain the code you're referring to and how it helps demonstrate this concept.

General Relationship Between Mean, Median, and Skewness
Right Skewness (Positive Skewness):

In a right-skewed distribution (also called positively skewed), the tail on the right side of the distribution is longer or fatter.
In such cases, mean > median.
This happens because the larger values (in the right tail) pull the mean to the right, while the median (which is the middle value) remains more central to the data.
Left Skewness (Negative Skewness):

In a left-skewed distribution (also called negatively skewed), the tail on the left side is longer or fatter.
In such cases, mean < median.
Here, smaller values (in the left tail) pull the mean to the left, while the median remains more central.
Symmetric Distribution:

In a symmetric distribution (e.g., a normal distribution), the mean = median because the data is evenly distributed on both sides.
Code Example to Demonstrate Mean, Median, and Skewness
Now, let's explore how the following code works and extend it to show the relationship between mean, median, and skewness:

python

import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Generate right-skewed data
data_right_skewed = np.random.exponential(scale=2, size=1000)

# Generate left-skewed data by taking the negative of an exponential distribution
data_left_skewed = -np.random.exponential(scale=2, size=1000)

# Generate normally distributed (symmetric) data
data_normal = np.random.normal(loc=0, scale=1, size=1000)

# Plot distributions
fig, axes = plt.subplots(1, 3, figsize=(18, 6))

sns.histplot(data_right_skewed, kde=True, ax=axes[0])
axes[0].set_title('Right Skewed Distribution')
axes[0].axvline(np.mean(data_right_skewed), color='r', linestyle='--', label=f'Mean: {np.mean(data_right_skewed):.2f}')
axes[0].axvline(np.median(data_right_skewed), color='g', linestyle='-', label=f'Median: {np.median(data_right_skewed):.2f}')
axes[0].legend()

sns.histplot(data_left_skewed, kde=True, ax=axes[1])
axes[1].set_title('Left Skewed Distribution')
axes[1].axvline(np.mean(data_left_skewed), color='r', linestyle='--', label=f'Mean: {np.mean(data_left_skewed):.2f}')
axes[1].axvline(np.median(data_left_skewed), color='g', linestyle='-', label=f'Median: {np.median(data_left_skewed):.2f}')
axes[1].legend()

sns.histplot(data_normal, kde=True, ax=axes[2])
axes[2].set_title('Normal Distribution (Symmetric)')
axes[2].axvline(np.mean(data_normal), color='r', linestyle='--', label=f'Mean: {np.mean(data_normal):.2f}')
axes[2].axvline(np.median(data_normal), color='g', linestyle='-', label=f'Median: {np.median(data_normal):.2f}')
axes[2].legend()

plt.tight_layout()
plt.show()
How the Code Works:
Data Generation:
The code generates three different datasets:
Right-Skewed Data: Generated using an exponential distribution (np.random.exponential), which naturally produces right skew.
Left-Skewed Data: Created by taking the negative of an exponential distribution (-np.random.exponential), making the data skew to the left.
Normally Distributed Data: Generated using the normal distribution (np.random.normal), which is symmetric by nature.
Plotting:
The code uses Seaborn’s histplot to create histograms with KDE (kernel density estimation) for each dataset. The red dashed line represents the mean, and the green solid line represents the median for each plot.
Visualizing Skewness:
The resulting histograms show how the relationship between mean and median varies:
Right-Skewed Distribution: The mean will be pulled toward the right tail, while the median stays more central.
Left-Skewed Distribution: The mean will be pulled toward the left tail, while the median stays closer to the center.
Symmetric Distribution: The mean and median will be very close or identical.
Explaining the Relationship Using Code
Let’s extend this explanation by discussing how outliers and the shape of the data pull the mean in one direction, while the median is less affected.

Right Skewness: The long tail on the right contains outliers, which heavily affect the mean because it's sensitive to extreme values. However, since the median only depends on the position of the middle value, it is less affected by these outliers. This causes the mean to be larger than the median.

Left Skewness: The long tail on the left contains smaller outliers, pulling the mean toward the left (making it smaller). The median again remains near the center, so the mean ends up smaller than the median.

Symmetry: Since both sides of the distribution are balanced, the mean and median overlap.

```
